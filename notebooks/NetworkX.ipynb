```python
import networkx as nx
print(nx.__version__)
```

## 无向图

```python
# 创建一个无向图
G = nx.Graph()

# 添加节点
G.add_node(1)
G.add_nodes_from(['A', 'B', 'C'])
# 添加边
G.add_edge(1, 'A')
G.add_edges_from([('A', 'B'), ('B', 'C')])

# 打印图的信息
print("Nodes:", G.nodes())
print("Edges:", G.edges())

# 可视化图
nx.draw(G, with_labels=True,pos=nx.spring_layout(G)) 
# pos 作用是指定节点布局方式，这里使用的是弹簧布局（spring layout），它会根据节点之间的连接关系自动调整节点的位置，使得图形看起来更美观。

```

```python
print(nx.is_directed(G))  # 检查图是否为有向图
print(nx.is_weighted(G))  # 检查图是否为加权图
```
